```python
# Assignment 6 Weather analysis - Knock Airport
# This notebook loads hourly weather data and creates temperature plots
# Libraries used
# - Pandas for data handling
# - Matplotlib for plotting charts

import pandas as pd 
import matplotlib.pyplot as plt

# This displays the plots inside the notebook
%matplotlib inline

# Load the knock airport weather data
# The first 22 lines contain station info so these lines are skipped

url = "https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv"

df = pd.read_csv(url, skiprows = 22)

# Show the first few rows and column names
df.head()
df.columns
```

```python
# Convert the 'date' column to dataframe objects
df["date"] = pd.to_datetime(df["date"])

# Set the 'date' as the index
df = df.set_index("date")

# View the data again
df.head()
```

```python
# Plot the hourly air temperature
# This gives a better idea of how the temperature changes 
plt.figure(figsize=(12, 4))

df["temp"].plot()

plt.title("Hourly Temperature at knock airport")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.grid(True)

plt.show
```

```python
# Calculate the mean temperature for each day
# The 'D' tells pandas to resample daily
daily_mean_temp = df["temp"].resample("D").mean()

plt.figure(figsize=(12, 4))

daily_mean_temp.plot()

plt.title("Daily Mean Temperature at Knock Airport")
plt.xlabel("Date")
plt.ylabel("Mean Temperature (C)")
plt.grid(True)

plt.show()
```

```python
# Calculate the mean temperature for each month
# The 'M' groups all temperature values by month
monthly_mean_temp = df["temp"].resample("M").mean()

plt.figure(figsize=(12, 4))

monthly_mean_temp.plot(marker="o")

plt.title("Monthly Mean Temperature at Knock Airport")
plt.xlabel("Month")
plt.ylabel("Mean Temperature (C)")
plt.grid(True)

plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

%matplotlib inline

# Load the knock airport weather data
url = "https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv"

df = pd.read_csv(url, skiprows = 22)

# Show the first few rows and column names
df.head()
df.columns
```

```python
# Convert the 'date' column to dataframe objects
df["date"] = pd.to_datetime(df["date"])

# Set the 'date' as the index
df = df.set_index("date")

df.head()
```

```python
plt.figure(figsize=(12, 4))

df["temp"].plot()

plt.title("Hourly Temperature at knock airport")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.grid(True)

plt.show
```
